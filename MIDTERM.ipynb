```python
#1.2
Companies = [MPC, MSFT, XRAY, KEYS]

for company in Companies: 
    print(company.name + ":")
    print(company["Adj Close"].agg([min, max]))
```

```python
#1.3
for company in Companies:
    company['log_returns'] = np.log(company['Adj Close'].astype('float64') / company['Adj Close'].astype('float64').shift(1))
    print(company.name + ':')
    print(company['log_returns'].agg(['min', 'max', 'median', 'mean']))
```
